```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
url = '/Users/amirrezakamkar/Downloads/amz_uk_price_prediction_dataset.csv'
df = pd.read_csv(url)
```

```python
df.columns
```

```python
#1:
category_frequency = df['category'].value_counts()
category_frequency
```

```python
top_categories = category_frequency.head()
```

```python
sns.barplot(y=top_categories.index, x=top_categories.values, palette="Set3");
```

```python
plt.figure(figsize=(16, 16))
top_categories.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"));
```

```python
#2:
price_mean = df['price'].mean()
price_median = df['price'].median()
price_mode = df['price'].mode()[0]
```

```python
price_mean, price_median, price_mode
```

```python
#the average is much higher than mean and mode, which shows there are some very expensive items.
```

```python
price_var = df['price'].var()
price_std = df['price'].std()
price_range = df['price'].max() - df['price'].min()
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
price_var, price_std, price_range, IQR
```

```python
# there are different prices and the variance is high, and there are also some very expensive products.
```

```python
sns.histplot(df['price'], kde=True, bins=10, color="salmon");
```

```python
# As most of the products are in a small price range, the histogram doesn't show some useful insights.
```

```python
#3:
rating_mean = df['stars'].mean()
rating_median= df['stars'].median()
rating_mode = df['stars'].mode()[0]
rating_mean, rating_median, rating_mode
```

```python
#most of them doesn't have any rating (0)
```

```python
rating_var = df['stars'].var()
rating_std = df['stars'].std()
Q1 = df['stars'].quantile(0.25)
Q3 = df['stars'].quantile(0.75)
IQR = Q3 - Q1
rating_var, rating_std, IQR
```

```python
#there is a wide variation.
```

```python
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurtosis()
skewness_rating, kurtosis_rating
```

```python
#they are not normally distributed, they lean toward lower values.
```
